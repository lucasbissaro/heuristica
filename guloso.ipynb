```python
# metaheuristica procurar ponto mais proximo de 2 pontos e colocar ele ali no meio

import math
import copy
def solucaoFinal(vRota, vItens, matriz, itens, vMin, vMax, v, maxTime, w):
    vRota = [0] + vRota + [len(matriz)-1]
    peso = 0
    valor=0
    itensCid=[]
    for x in matriz:
        itensCid.append([])
    for x in vItens:        
        #print(x)
        itensCid[int(itens[x][3])].append(itens[x])
        peso += itens[x][2]
        valor += itens[x][1]
    
    t = 0
    pesoAtual = 0
    for index, x in enumerate(vRota):
        #print(index,x , len(vRota))
        if(index < len(vRota)-1):
            for y in itensCid[x]:
                #print('cidade', x-1, 'peso', pesoAtual)
                pesoAtual +=y[2]
            #print(matriz[x-1][vRota[index+1]-1], vMax , v*pesoAtual, vMin)
            t += matriz[x][vRota[index+1]]/max(vMax - v*pesoAtual, vMin)
    tempoAtual = (t/maxTime) * (pesoAtual/w)
    print('tempo', t,maxTime, pesoAtual, w, 'peso:', tempoAtual, valor)
    #print(t,maxTime , peso,w)
    if(t<=maxTime and pesoAtual<=w):
        return valor
    else:
        return -1
    return 0
def solucaoFinal2(vRota, vItens, matriz, itens, vMin, vMax, v, maxTime, w):
    vRota = [0] + vRota + [len(matriz)-1]
    peso = 0
    valor=0
    itensCid=[]
    for x in matriz:
        itensCid.append([])
    for x in vItens:        
        #print(x)
        itensCid[int(itens[x][3])].append(itens[x])
        peso += itens[x][2]
        valor += itens[x][1]
    
    t = 0
    pesoAtual = 0
    for index, x in enumerate(vRota):
        #print(index,x , len(vRota))
        if(index < len(vRota)-1):
            for y in itensCid[x]:
                #print('cidade', x-1, 'peso', pesoAtual)
                pesoAtual +=y[2]
            #print(matriz[x-1][vRota[index+1]-1], vMax , v*pesoAtual, vMin)
            t += matriz[x][vRota[index+1]]/max(vMax - v*pesoAtual, vMin)
    tempoAtual = (t/maxTime) * (pesoAtual/w)
    print('tempo', t,maxTime, pesoAtual, w, 'peso:', tempoAtual)
    #print(t,maxTime , peso,w)
    if(t<=maxTime and pesoAtual<=w):
        return valor, tempoAtual
    else:
        return -1, tempoAtual
    return 0
def dist(a, b):
    #print(a)
    #print(b)
    t = math.sqrt((a[1]-b[1])*(a[1]-b[1])+(a[2]-b[2])*(a[2]-b[2]))
    if int(t)==t:
        n = int(t)
    else:
        n = int(t)+1
    return n

def solucao(vRota, cidadeItens, matriz):
    vRota = [0] + vRota + [len(matriz)-1]
    peso = 0
    valor=0
    v = (maxSpeed - minSpeed)/capacity
    t = 0
    pesoAtual = 0
    for index, x in enumerate(vRota):
        #print(index,x , len(vRota))
        if(index < len(vRota)-1):
            valor+=cidadeItens[x]['valor']    
            pesoAtual +=cidadeItens[x]['peso']            
            t += matriz[x][vRota[index+1]]/max(maxSpeed - v*pesoAtual, minSpeed)
    if(t<=maxTime and pesoAtual<=capacity):
        return valor
    else:
        return -1
    return 0


def proximaCidade(matriz, dadosCidade, cidadeAtual, cidadeVisitada):
    maxi = -1
    ind = -1
    for index, x in enumerate(dadosCidade):
        if  matriz[cidadeAtual][index]==0 and cidadeAtual!=index:
            print(cidadeAtual, index, matriz[cidadeAtual][index])
        if (cidadeAtual!=index and maxi<(x/matriz[cidadeAtual][index]) and not cidadeVisitada[index]):            
            maxi = x/matriz[cidadeAtual][index]
            ind = index            
            
    return ind


def adicionaMaximoItens(rota, cidadeItens, cidadeAtual, itens, matriz, todosItens):
    rotaCopia = copy.deepcopy(rota)
    rotaCopia.append(cidadeAtual)
    lastResult = -1
    for x in itens:       
        cidadeItens[cidadeAtual]['valor']+=x[1]
        cidadeItens[cidadeAtual]['peso']+=x[2]         
        cidadeItens[cidadeAtual]['qtdItens']+=1
        todosItens.append(x[0])
        result = solucao(rotaCopia, cidadeItens, matriz)
        #print(rotaCopia, cidadeItens[cidadeAtual])
        if result==-1:            
            cidadeItens[cidadeAtual]['valor']-=x[1]
            cidadeItens[cidadeAtual]['peso']-=x[2]
            cidadeItens[cidadeAtual]['qtdItens']-=1
            todosItens.remove(x[0])
            break
        else:
            lastResult = result
            if(cidadeAtual not in rota):
                rota.append(cidadeAtual)
            
    return lastResult
```
